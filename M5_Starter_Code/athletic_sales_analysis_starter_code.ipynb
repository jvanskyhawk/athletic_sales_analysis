```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df_2020 = pd.read_csv('./Resources/athletic_sales_2020.csv')
df_2021 = pd.read_csv('./Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
df_2020.head()
```

```python
# Display the 2021 sales DataFrame
df_2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
df_2020.dtypes
```

```python
# Check the 2021 sales data types.
df_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
df = pd.concat([df_2020, df_2021], axis='rows', join='inner', ignore_index=True)
df.head()
```

```python
# Check if any values are null.
df.isnull().sum()
```

```python
# Check the data type of each column
df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
df['invoice_date'] = pd.to_datetime(df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
total_df = df.groupby(['region', 'state', 'city'])[['units_sold']].sum()
total_df = total_df.rename(columns={'units_sold':'Total_Products_Sold'})


# Show the top 5 results.
total_df = total_df.sort_values(by='Total_Products_Sold', ascending=False)
total_df.head()                   
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
df_pivot = pd.pivot_table(df, values='units_sold', index=['region','state','city'], aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
df_pivot = df_pivot.rename(columns={'units_sold':'Total_Products_Sold'})

# Show the top 5 results.
df_pivot = df_pivot.sort_values(by='Total_Products_Sold', ascending=False).head()
df_pivot.head()
```

### 3. Determine which Region had the Most Sales

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
df['product'].value_counts()
```

```python
# Filter the sales data to get the women's athletic footwear sales data.
footwear = df[df["product"].str.contains("Women's Athletic Footwear")]
footwear.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
footwear_df = footwear.groupby(['retailer', 'region', 'state', 'city'])[['units_sold']].sum()
footwear_df = footwear_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
footwear_df = footwear_df.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
footwear_df.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
footwear_pivot = pd.pivot_table(footwear, values='units_sold', index=['retailer', 'region','state','city'], aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
footwear_pivot = footwear_pivot.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
footwear_pivot = footwear_pivot.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
footwear_pivot.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
invoice_pivot = pd.pivot_table(df, values='total_sales', index=['invoice_date'], aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
invoice_pivot = invoice_pivot.rename(columns={'total_sales': 'Total Sales'})

# Show the table.
invoice_pivot.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_pivot = invoice_pivot.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_pivot = daily_pivot.sort_values(by='Total Sales', ascending=False)
daily_pivot.head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_pivot = invoice_pivot.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_pivot = weekly_pivot.sort_values(by='Total Sales', ascending=False)
weekly_pivot.head(10)
```
